## Bivariate analysis

## linear regression model

```python
from sklearn.linear_model import LinearRegression, LogisticRegression

X_lin = df.drop('Outcome',axis=1)
y_lin = df['Outcome']

lin = LinearRegression()
lin.fit(X_lin, y_lin)
    
# Make predictions
pred_linear = lin.predict(X_lin)
pred_linear
```

## logistic regression model

```python
# logistic model
X_log = df.drop('Outcome',axis=1)
y_log = df['Outcome']

# Fit the logistic regression model
log = LogisticRegression()
log.fit(X_log, y_log)

# Make predictions
pred_log = log.predict(X_log)
pred_log
```

## Multiple Regression analysis

```python
import pandas as pd
from statsmodels.formula.api import ols

# Assuming the columns include features and the target variable 'Outcome'
# Replace 'Outcome' with the actual target column name in your dataset
outcome = 'Outcome'

# Perform multiple linear regression
model = ols(f'{outcome} ~ Pregnancies + Glucose + BloodPressure + SkinThickness + Insulin + BMI + DiabetesPedigreeFunction + Age', data=df).fit()

# Print the summary of the regression model
print(model.summary())
```
